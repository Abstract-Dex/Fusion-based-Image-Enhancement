```python
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from PIL import Image

im = Image.open("images/bio.jpg")
px = im.load()

ax = plt.axes(projection = '3d')

x = []
y = []
z = []
c = []

for row in range(0,im.height):
       for col in range(0, im.width):
            pix = px[col,row]
            x.append(pix[0])
            y.append(pix[1])
            z.append(pix[2])
            c.append((pix[0] / 255, pix[1] / 255, pix[2] / 255))

for row in range(0,im.height):
     for col in range(0, im.width):
         pix = px[col,row]
         newCol = (pix[0] / 255, pix[1] / 255, pix[2] / 255)

         if(not newCol in c):
            x.append(pix[0])
            y.append(pix[1])
            z.append(pix[2])
            c.append(newCol)

ax.scatter(x,y,z, c = c)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

im = cv2.imread("paper results/brackish4_mmclahe.jpg")
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

ax = plt.axes(projection='3d')

x = []
y = []
z = []
unique_colors = []

for row in range(im.shape[0]):
    for col in range(im.shape[1]):
        pix = im[row, col]
        x.append(pix[0])
        y.append(pix[1])
        z.append(pix[2])
        unique_colors.append(tuple(pix / 255.0))

ax.scatter(x, y, z, c=unique_colors)
plt.show()
# plt.savefig('color models/brackish1.png')
```
